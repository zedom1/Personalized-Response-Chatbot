```python
import re
import codecs
```

```python
txt = codecs.open('./data/seq2seq/sentences.txt',"r","utf-8").read()
```

```python
exec(open("preppy.py").read())
```

```python
split_reg = re.compile('\n')
```

# parse the raw text of QA into dictionary

```python
sentences = split_reg.split(txt)
```

```python
print(sentences[0])
```

One single line demo

```python
verses_splitter = re.compile('(?P<sentence>.+)',)
```

```python
gen = verses_splitter.finditer(sentences[0].replace("\n",""))
```

```python
next(gen).groupdict()
```

Lets run all of that on the entire text

```python
examples= []
for line in sentences:
    for verse_regex_match in verses_splitter.finditer(line):
        example = verse_regex_match.groupdict()
        example.update({"sentence":example["sentence"].strip()})
        examples.append(example)
```

```python
len(examples)
print((examples[0]))
```

```python
import pickle
pickle.dump(examples,open('./data/seq2seq/examples.pkl','wb'))
```

## Make them into TF records

```python
import tensorflow as tf
from preppy import Preppy
import numpy as np
```

```python
import random
random.shuffle(examples)
val_size = int(len(examples)*0.25)
val,train = examples[:val_size], examples[val_size:]
```

```python
def tokenizer(sentence):
    return sentence.split()
    
BP =Preppy(tokenizer_fn=tokenizer) #Charecter level tokenization
BP.init_vocab()

for example in examples:
    BP.prepare_vocab(sequence=example["sentence"])
    
BP.vocab_filter()

for (data,path) in [(val,'./data/seq2seq/val.tfrecord'),(train,'./data/seq2seq/train.tfrecord')]:
    with open(path,'w') as f:
        writer = tf.python_io.TFRecordWriter(f.name)
    for example in data:
        record = BP.sequence_to_tf_example(sequence=example["sentence"])
        writer.write(record.SerializeToString())
```

```python
BP.update_reverse_vocab()
```

```python
pickle.dump(BP,open('./data/seq2seq/preppy.pkl','wb'))
```
